# Distribucion aleatoria en una esfera

Genera las contenedoras.

```python
n = 5000
x = np.zeros(n)
y = np.zeros(n)
z = np.zeros(n)
```

Genera numeros aleatorios y llena las contenedoras solo con los que se encuentran dentro de la esfera.

```python
l = 0
while l < n :
    i = 1 - 2*np.random.random(1)
    j = 1 - 2*np.random.random(1)
    k = 1 - 2*np.random.random(1)
    if np.sqrt(i**2 + j**2 + k**2) < 1 :
        x[l] = i
        y[l] = j
        z[l] = k
        l = l+1
```

### Grafica bidimensional

```python
ind = where((z < 1))
scatter(x[ind], y[ind], alpha = 0.1)
```

###  Grafica tridimensional

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c= 'b', marker='o', alpha = 0.3)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
